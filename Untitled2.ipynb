Найдите минимальный возраст учащегося школы Mousinho da Silveira. Вывести

```python
min_age_mousinho = data[data['school'] == 'MS']['age'].min()
students_min_age_mousinho = data[(data['school'] == 'MS') & (data['age'] == min_age_mousinho)]
print("Minimum age:", min_age_mousinho)
students_min_age_mousinho
```

Найдите разность между средними итоговыми оценками студентов, состоящих и не
состоящих в романтических отношениях. В качестве ответа приведите число, округленное
до двух значащих цифр после запятой. Вывести информацию об этих студентах

```python
mean_grade_in_relationship = data[data['romantic'] == 'yes']['G3'].mean()
mean_grade_not_in_relationship = data[data['romantic'] == 'no']['G3'].mean()
difference = mean_grade_in_relationship - mean_grade_not_in_relationship
rounded_difference = round(difference, 2)

print("Разность между средними оценками студентов, состоящих и не состоящих в романтических отношениях:", rounded_difference)
```

Найти максимальное количество пропусков у студента. Вывести информацию об

Найти максимальный возраст учащегося, имеющего более 5 пропусков. Вывести его.
Вывести информацию об этих студентах.

```python
max_age_with_high_absences = data[data['absences'] > 5]['age'].max()
students_with_max_age_and_high_absences = data[(data['absences'] > 5) & (data['age'] == max_age_with_high_absences)]
print("Максимальный возраст учащегося с более чем 5 пропусками:", max_age_with_high_absences)
print("Информация о студентах:")
students_with_max_age_and_high_absences
```

 Найдите разность между средними итоговыми оценками студентов, у которых есть
домашний интернет и у которых его нет. В качестве ответа приведите число, округленное
до двух значащих цифр после запятой. Вывести информацию об этих студентах

```python
mean_grade_with_internet = data[data['internet'] == 'yes']['G3'].mean()
mean_grade_without_internet = data[data['internet'] == 'no']['G3'].mean()

difference = mean_grade_with_internet - mean_grade_without_internet
rounded_difference = round(difference, 2)

print("Разность между средними оценками студентов с домашним интернетом и без него:", rounded_difference)
```
